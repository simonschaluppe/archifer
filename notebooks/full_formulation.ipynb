```python
import pulp
```

```python
prob = pulp.LpProblem("Full_Formulation", pulp.LpMinimize)
prob
```

### Create our decision variables

```python
var_names = ["chicken", "Beef", "Mutton", "Rice", "Wheat", "Gel"]
```

```python
vars = pulp.LpVariable.dict("Ingredients", 
                            var_names, 0, 100, pulp.LpContinuous)
vars
```

```python
x1 = pulp.LpVariable("chicken", 0, 100, pulp.LpInteger)
x2 = pulp.LpVariable("Beef", 0, 100, pulp.LpInteger)
x3 = pulp.LpVariable("Mutton", 0, 100, pulp.LpInteger)
x4 = pulp.LpVariable("Rice", 0, 100, pulp.LpInteger)
x5 = pulp.LpVariable("Wheat", 0, 100, pulp.LpInteger)
x6 = pulp.LpVariable("Gel", 0, 100, pulp.LpInteger)
```

```python
prob += 0.013*x1 + 0.016*x2 + 0.010*x3 + 0.002*x4 + 0.005*x5 + 0.001*x6
prob
```

```python
prob += x1 + x2 + x3 + x4 + x5 + x6 == 100, "Total Content"
prob
```

```python
prob += 0.1*x1 + 0.2*x2 + 0.1*x3 + 0*x4 + 0.4*x5 + 0 * x6 >= 8, "Protein"
prob
```

```python
prob += 0.08*x1 + 0.1*x2 + 0.011*x3 +0.01*x4 +0.01*x5 +0*x6 >= 6, "fat"
prob
```

```python
prob += 0.001*x1 + 0.005*x2 + 0.003 * x3 + 0.100 *x4 + 0.15*x5 + 0*x6 <= 2, "fibre"
prob += 0.002*x1 + 0.005*x2 + 0.007*x3 + 0.002*x4 + 0.008*x5 + 0*x6 <= 0.4, "salt"
```

```python
prob
```

```python
prob.solve()
```

```python
clean = pulp.LpProblem("Clean_Formulation", pulp.LpMinimize)
Ingredients = ["CHICKEN", "BEEF", "MUTTON", "RICE", "WHEAT", "GEL"]
costs = {
    "CHICKEN": 0.013,
    "BEEF": 0.008,
    "MUTTON": 0.010,
    "RICE": 0.002,
    "WHEAT": 0.005,
    "GEL": 0.001,
}

# A dictionary of the protein percent in each of the Ingredients is created
proteinPercent = {
    "CHICKEN": 0.100,
    "BEEF": 0.200,
    "MUTTON": 0.150,
    "RICE": 0.000,
    "WHEAT": 0.040,
    "GEL": 0.000,
}

# A dictionary of the fat percent in each of the Ingredients is created
fatPercent = {
    "CHICKEN": 0.080,
    "BEEF": 0.100,
    "MUTTON": 0.110,
    "RICE": 0.010,
    "WHEAT": 0.010,
    "GEL": 0.000,
}

# A dictionary of the fibre percent in each of the Ingredients is created
fibrePercent = {
    "CHICKEN": 0.001,
    "BEEF": 0.005,
    "MUTTON": 0.003,
    "RICE": 0.100,
    "WHEAT": 0.150,
    "GEL": 0.000,
}

# A dictionary of the salt percent in each of the Ingredients is created
saltPercent = {
    "CHICKEN": 0.002,
    "BEEF": 0.005,
    "MUTTON": 0.007,
    "RICE": 0.002,
    "WHEAT": 0.008,
    "GEL": 0.000,
}
```

```python
for key, value in saltPercent.items():
    print(key, value)
```

```python
ing_vars = pulp.LpVariable.dicts("Ingredients", Ingredients, 0, None, pulp.LpContinuous)
clean.variables()
```

```python
clean += pulp.lpSum([costs[i] * ing_vars[i] for i in Ingredients])
```

```python
clean += pulp.lpSum([ing_vars[i] for i in Ingredients]) == 100
```

```python
clean += pulp.lpSum([proteinPercent[i] * ing_vars[i] for i in Ingredients]) >= 8
clean += pulp.lpSum([fatPercent[i] * ing_vars[i] for i in Ingredients]) >= 6
clean += pulp.lpSum([fibrePercent[i] * ing_vars[i] for i in Ingredients]) <= 2
clean += pulp.lpSum([saltPercent[i] * ing_vars[i] for i in Ingredients]) <= 0.4
```

```python
clean
```

```python
clean.solve()
```

```python
for ing in clean.variables():
    print(ing.name, ing.varValue)
```

```python
sum(ing.varValue * costs[ing.name] for ing in clean.variables())
```
